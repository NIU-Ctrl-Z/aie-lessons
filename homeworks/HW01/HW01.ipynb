**2. Примеры создания массивов:**
- из обычных списков;
- с использованием функций `zeros`, `ones`, `arange` или аналогичных.

```python
arr_list = np.array([1, 2, 3, 4, 5]) # Создание массива из обычного списка
arr_list
```

```python
arr_zeros = np.zeros((5, 5)) # Создание двумерного массива из нулей размером 5 на 5
arr_zeros
```

```python
arr_ones = np.ones(3) # Создание одномерного массива из единиц длиной 3 элемента
arr_ones
```

```python
arr_range = np.arange(1, 20, 4) # Создание массива с использованием range
arr_range
```
